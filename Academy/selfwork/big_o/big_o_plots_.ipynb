```python
import matplotlib.pyplot as plt
```

```python
steps = []

def constant(n):
    return 1

for i in range(1, 100):
    steps.append(constant(i))
plt.plot(steps)
```

```python
steps = []

def linear(n):
    return n

for i in range(1, 100):
    steps.append(linear(i))

plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
steps = []

def linear(n):
    return 2*n

for i in range(1, 100):
    steps.append(linear(i))

plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
steps = []

def quadratic_algo(items):
    s = []
    for item in items:
        for item2 in items:
            s.append(item + item2)
    return s


for i in range(1, 100):
    l = list(range(i))
    steps.append(len(quadratic_algo(l)))

plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
steps = []

def search_algo(num, items):
    s = []
    for item in items:
        if item == num:
            s.append(item)
        else:
            s.append(None)
    return s

for i in range(1, 100):
    l = list(range(i))
    steps.append(len(search_algo(10, l)))

plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
steps = []

def return_squares(n):
    square_list = []
    for num in n:
        square_list.append(num * num)

    return square_list

for i in range(1, 100):
    l = list(range(i))
    steps.append(len(return_squares(l)))

plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```
