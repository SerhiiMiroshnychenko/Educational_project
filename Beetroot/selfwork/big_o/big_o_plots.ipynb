```python
import matplotlib.pyplot as plt
```

```python
from typing import List, Tuple, Any
```

```python
from random import randint
```

```python
# We assume that all lists passed to functions are the same length N
```

```python
def question1(first_list: List[int], second_list: List[int]) -> List[int]: # n
    res: List[int] = []
    for el_first_list in first_list:
        if el_first_list in second_list:
            res.append(el_first_list)
    return res
```

```python
def question1_big_o(first_list: List[int], second_list: List[int]) -> tuple[List[int], List[int]]: # n
    s: List[int] = [] # Big O
    res: List[int] = []
    for el_first_list in first_list:
        if el_first_list in second_list:
            res.append(el_first_list)
            s.append(1) # Big O

    return res, s

steps = []
for i in range(1, 1000):
    list1 = []
    list2 = []
    for _ in range(i):
        list1.append(randint(0, 10))
        list2.append(randint(0, 10))
    steps.append(len(question1_big_o(list1, list2)[1]))


plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
def question2(n: int) -> int:
	for _ in range(10):
		n **= 3
	return n
```

```python
def question2_big_o(n: int) -> Tuple[int, list]:
    s: List[int] = []
    for _ in range(10):
        n **= 3
        s.append(1)
    return n, s

steps = []
for _ in range(1, 1000):
    n = randint(0, 10)
    steps.append(len(question2_big_o(n)[1]))


plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
def question3(first_list: List[int], second_list: List[int])-> List[int]: # n*n
   temp: List[int] = first_list[:]
   for el_second_list in second_list:
      flag = False
      for check in temp:
         if el_second_list == check:
            flag = True
            break
      if not flag:
         temp.append(second_list)
   return temp
```

```python
def question3_big_o(first_list: List[int], second_list: List[int])-> Tuple[List[int], List[int]]: # n*n
   s: List[int] = []
   temp: List[int] = first_list[:]
   for el_second_list in second_list:
      flag = False
      for check in temp:
          s.append(1)
          if el_second_list == check:
              flag = True
              break
      if not flag:
         temp.append(second_list)
   return temp, s

steps = []
for i in range(1, 100):
    list1 = []
    list2 = []
    for _ in range(i):
        list1.append(randint(0, 1000))
        list2.append(randint(0, 1000))
    steps.append(len(question3_big_o(list1, list2)[1]))


plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
def question4(input_list: List[int]) -> int: # n
  res: int = 0
  for el in input_list:
    if el > res:
      res = el
  return res
```

```python
def question4_big_o(input_list: List[int]) -> tuple[int, list]: # n
  s: List[int] = []
  res: int = 0
  for el in input_list:
    s.append(1)
    if el > res:
      res = el
  return res, s

steps = []
for i in range(1, 100):
    list1 = []
    for _ in range(i):
        list1.append(randint(0, 1000))
    steps.append(len(question4_big_o(list1)[1]))



plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
def question5(n: int) -> List[Tuple[int, int]]: # n*n
    res: List[Tuple[int, int]] = []
    for i in range(n):
        for j in range(n):
            res.append((i, j))
    return res
```

```python
def question5_big_o(n: int) -> Tuple[List[Tuple[int, int]], List[int]]: # n*n
    s: List[int] = []
    res: List[Tuple[int, int]] = []
    for i in range(n):
        for j in range(n):
            s.append(1)
            res.append((i, j))
    return res, s

steps = []
for n in range(1, 100):
    steps.append(len(question5_big_o(n)[1]))

plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
def question6(n: int) -> int: # log n
    while n > 1:
        n /= 2
    return n
```

```python
def question6_big_o(n: int) -> tuple[int, list]: # log n
    s: List[int] = []
    while n > 1:
        n /= 2
        s.append(1)
    return n, s

steps = []
for n in range(1, 10000):
    steps.append(len(question6_big_o(n)[1]))

plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
steps = []

def quadratic_algo(items):
    s = []
    for item in items:
        for item2 in items:
            s.append(item + item2)
    return s


for i in range(1, 100):
    l = list(range(i))
    steps.append(len(quadratic_algo(l)))

plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
steps = []

def search_algo(num, items):
    s = []
    for item in items:
        if item == num:
            s.append(item)
        else:
            s.append(None)
    return s

for i in range(1, 100):
    l = list(range(i))
    steps.append(len(search_algo(10, l)))

plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
steps = []

def return_squares(n):
    square_list = []
    for num in n:
        square_list.append(num * num)

    return square_list

for i in range(1, 100):
    l = list(range(i))
    steps.append(len(return_squares(l)))

plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```
