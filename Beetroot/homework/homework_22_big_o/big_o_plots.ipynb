```python
import matplotlib.pyplot as plt
```

```python
from typing import List, Tuple
```

```python
from random import randint
```

# We assume that all lists passed to functions are the same length N

```python
# Функція 1
def question1(first_list: List[int], second_list: List[int]) -> List[int]: # n
    res: List[int] = []
    for el_first_list in first_list: # O(n)
        if el_first_list in second_list: # O(n)
            res.append(el_first_list) # O(1)
    return res
# O(n) + O(n) + O(1) = 2*O(n) + 0(1) = O(n)
```

```python
# Модифікуємо для визначення Big O
def question1_big_o(first_list: List[int], second_list: List[int]) -> tuple[List[int], List[int]]: # n
    s: List[int] = [] # Big O
    res: List[int] = []
    for el_first_list in first_list:
        if el_first_list in second_list:
            res.append(el_first_list)
            s.append(1) # Big O

    return res, s

steps = []
for i in range(1, 1000):
    list1 = []
    list2 = []
    for _ in range(i):
        list1.append(randint(0, 10))
        list2.append(randint(0, 10))
    steps.append(len(question1_big_o(list1, list2)[1]))


plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
# Функція 2
def question2(n: int) -> int:
	for _ in range(10): # O(1)
		n **= 3
	return n
# O(1)
```

```python
# Модифікуємо для визначення Big O
def question2_big_o(n: int) -> Tuple[int, list]:
    s: List[int] = []
    for _ in range(10):
        n **= 3
        s.append(1)
    return n, s

steps = []
for _ in range(1, 1000):
    n = randint(0, 10)
    steps.append(len(question2_big_o(n)[1]))


plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
# Функція 3
def question3(first_list: List[int], second_list: List[int])-> List[int]: # n*n
   temp: List[int] = first_list[:] # O(n)
   for el_second_list in second_list: # O(n)
      flag = False
      for check in temp: # O(n)
         if el_second_list == check: # O(1)
            flag = True
            break
      if not flag: # O(1)
         temp.append(second_list) # O(1)
   return temp
# O(n) + O(n) * O(n) + O(1) + O(1) + O(1) = O(n) + O(n*n) + 3*O(1) = O(n*n)
```

```python
# Модифікуємо для визначення Big O
def question3_big_o(first_list: List[int], second_list: List[int])-> Tuple[List[int], List[int]]: # n*n
   s: List[int] = []
   temp: List[int] = first_list[:]
   for el_second_list in second_list:
      flag = False
      for check in temp:
          s.append(1)
          if el_second_list == check:
              flag = True
              break
      if not flag:
         temp.append(second_list)
   return temp, s

steps = []
for i in range(1, 100):
    list1 = []
    list2 = []
    for _ in range(i):
        list1.append(randint(0, 1000))
        list2.append(randint(0, 1000))
    steps.append(len(question3_big_o(list1, list2)[1]))


plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
# Функція 4
def question4(input_list: List[int]) -> int: # n
  res: int = 0
  for el in input_list: # O(n)
    if el > res: # O(1)
      res = el
  return res
# O(n) + O(1) = O(n)
```

```python
# Модифікуємо для визначення Big O
def question4_big_o(input_list: List[int]) -> tuple[int, list]: # n
  s: List[int] = []
  res: int = 0
  for el in input_list:
    s.append(1)
    if el > res:
      res = el
  return res, s

steps = []
for i in range(1, 100):
    list1 = [randint(0, 1000) for _ in range(i)]
    steps.append(len(question4_big_o(list1)[1]))

plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
# Функція 5
def question5(n: int) -> List[Tuple[int, int]]: # n*n
    res: List[Tuple[int, int]] = []
    for i in range(n): # O(n)
        for j in range(n): # O(n)
            res.append((i, j)) # O(1)
    return res
# O(n) * O(n) + O(1) = O(n*n) + O(1) = O(n*n)
```

```python
# Модифікуємо для визначення Big O
def question5_big_o(n: int) -> Tuple[List[Tuple[int, int]], List[int]]: # n*n
    s: List[int] = []
    res: List[Tuple[int, int]] = []
    for i in range(n):
        for j in range(n):
            s.append(1)
            res.append((i, j))
    return res, s

steps = [len(question5_big_o(n)[1]) for n in range(1, 100)]

print(steps)

plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```

```python
# Функція 6
def question6(n: int) -> int: # log n
    while n > 1: # O(log n)
        n /= 2
    return n
# O(log n)
```

```python
# Модифікуємо для визначення Big O
def question6_big_o(n: int) -> tuple[int, list]: # log n
    s: List[int] = []
    while n > 1:
        n /= 2
        s.append(1)
    return n, s

steps = [len(question6_big_o(n)[1]) for n in range(1, 1000)]

plt.plot(steps)
plt.xlabel('Inputs')
plt.ylabel('Steps')
```
